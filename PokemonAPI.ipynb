# Data Enrichment using API 
# Project Pikachu Data Enrichment

## 1. Import Libraries

```python
# import libraries 
import numpy as np
import pandas as pd
import requests
import json
import os
```

## 2. Use function requests to get API endpoint

```python
endpoint_get = 'https://api.pokemontcg.io/v2/cards'
r = requests.get(endpoint_get)
pokemon_dict = r.json()
```

## 3. Create Variable to save list pokemon name

```python
pokemon_dict
```

```python
pokemon_list = ['Ditto', 'Clefairy', 'Pikachu', 'Ninetales', 'Alakazam',
             'Bulbasaur', 'Charizard', 'Gardevoir','Mewtwo', 'Blastoise']
pokemon_list
```

```python
pokemon_data = []

pokemon_dict['data'][0]

for i in pokemon_list:
    for j in range(0, len(pokemon_dict['data'])):
        if pokemon_dict['data'][j]['name'] in i:
            pokemon_data.append(pokemon_dict['data'][j])
```

```python
df_pokemon = pd.DataFrame(pokemon_data)
df_pokemon
```

## 4. tidy up some columns written in Json 

### 4.1 weakness columns

```python
weakness = df_pokemon['weaknesses'].apply(pd.Series)
pokemon_weakness= weakness[0].apply(pd.Series)


weakness_dict = {'type': 'weaknesses type', 'value': 'weaknesses value'}
 
pokemon_weakness.rename(columns=weakness_dict, inplace=True)
for i in pokemon_weakness.columns:
    df_pokemon[i] = pokemon_weakness[i]
df_pokemon
```

### 4.2 abilities columns

```python
abilities = df_pokemon['abilities'].apply(pd.Series)
pokemon_abilities = abilities[0].apply(pd.Series)

abilities_dict = {'name': 'abilities name', 'text': 'abilities effect', 'type': 'pokemon type'} 
pokemon_abilities.rename(columns=abilities_dict, inplace=True)
del pokemon_abilities[0]

for i in pokemon_abilities.columns:
    df_pokemon[i] = pokemon_abilities[i]
    
df_pokemon
```

### 4.3 Resistances columns

```python
resistances = df_pokemon['resistances'].apply(pd.Series)
pokemon_resistances= resistances[0].apply(pd.Series)
resistances_dict = {'type': 'resistances type', 'value': 'resistances valus'}
 
pokemon_resistances.rename(columns=resistances_dict, inplace=True)
del pokemon_resistances[0]

for i in pokemon_resistances.columns:
    df_pokemon[i] = pokemon_resistances[i]

df_pokemon
```

### 4.4 Attacks columns

```python
attack= df_pokemon['attacks'].apply(pd.Series)
pokemon_attack = attack[0].apply(pd.Series)
attack_dict = {'name': 'attack name', 'cost': 'attack cost', 'convertedEnergyCost': 'converted energy cost', 'text': 'attack description'}
 
pokemon_attack.rename(columns=attack_dict, inplace=True)
del pokemon_attack[0]

for i in pokemon_attack.columns:
    df_pokemon[i] = pokemon_attack[i]

df_pokemon
```
